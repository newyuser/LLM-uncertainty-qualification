```python
import re
import os
import json
import numpy as np
import pandas as pd
def jsonl_to_df(file_name):
    file_path = f'D:/ollama_fact/data/black_box_data/LLM_train_modified_res/{file_name}.jsonl'
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            try: 
                data.append(json.loads(line))
            except json.decoder.JSONDecodeError as e:
                print(line)
    df = pd.DataFrame(data)
    return df
```

```python
file_name = "MMLU"
df = jsonl_to_df(file_name)
```

```python
# Manual modification
df.loc[22, ['A', 'B', 'C', 'D', 'E', 'F']] = [0,80,0,20,0,0]
```

```python
df.fillna(0, inplace=True)

def uniform_types(x):
    if isinstance(x, list):
        if len(x) > 0: x = x[0]
        else: x = 0
    elif isinstance(x, str): 
        try:
            x = int(x)
        except ValueError:
            print(x)
            x = 0
    return x
df[['A', 'B', 'C', 'D', 'E', 'F']] = df[['A', 'B', 'C', 'D', 'E', 'F']].applymap(uniform_types)

# df = df[df['choices'] != {"A": "True, True", "B": "True, False", "C": "False, True", "D": "False, False", "E": "I don't know", "F": "None of the above"}]
# file_name = file_name+'_w'
    
data = np.array(df[['A', 'B', 'C', 'D', 'E', 'F']])
data[data > 100] = 0
for i in range(data.shape[0]):
    j = 0
    while data[i].sum() > 100 and j < 6:
        data[i, np.argpartition(data[i], 1)[j]] = 0
        j += 1
df[['A', 'B', 'C', 'D', 'E', 'F']] = data
jsonl_lines = df.to_json(orient='records', lines=True)

# 将JSONL字符串写入到文件中
with open(f'D:/ollama_fact/data/black_box_data/LLM_train_modified_res/{file_name}.jsonl', 'w') as f:
    f.write(jsonl_lines)
```
